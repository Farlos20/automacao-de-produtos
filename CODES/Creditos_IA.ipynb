### Case: Score de Crédito dos Clientes

Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, Ok, Bom

Arquivos da aula: https://drive.google.com/drive/folders/1FbDqVq4XLvU85VBlVIMJ73p9oOu6u2-J?usp=drive_link

```python
#Importar e tratar banco de dados
import pandas as pd
tabela = pd.read_csv("clientes.csv")
tabela = tabela.drop(columns="id_cliente")
display(tabela)
display(tabela.info())

#Poor = ruim
#Stadard = OK
#Good = Bom



```

```python
#conversão das colunas de object para numeros :
#profissão
#mix_credito

from sklearn.preprocessing import LabelEncoder
codificador_01 = LabelEncoder()

tabela["profissao"] = codificador_01.fit_transform(tabela["profissao"])

codificador_02 = LabelEncoder()
tabela["mix_credito"] = codificador_02.fit_transform(tabela["mix_credito"])

codificador_04 = LabelEncoder()
tabela["comportamento_pagamento"] = codificador_04.fit_transform(tabela["comportamento_pagamento"])

#AGORA SEPARAR AS INFOR PARA A IA
#Y = TODA A COLUNA OU IDEIA CENTRAL Q DESEJA PREVIÇÂO,Q A IA ACERTE
#x = tODO O RESTO Q DARA PARA A IA APRENDER PARA PODER PREVER E RESPONDER
y = tabela["score_credito"]
x = tabela.drop(columns= "score_credito")

from sklearn.model_selection import train_test_split
x_treino,x_teste,y_treino,y_teste = train_test_split(x,y,test_size=0.3)

display(tabela.info())
```

```python
#Criando e adicionando a IA
from sklearn.ensemble import RandomForestClassifier
arvore_decisão = RandomForestClassifier()
arvore_decisão.fit(x_treino, y_treino)

teste_arvore = arvore_decisão.predict(x_teste)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_teste,teste_arvore))
```
